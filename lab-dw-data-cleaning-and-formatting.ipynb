## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
# Your code here
def categorize_gender(x):
    if isinstance(x, str):
        if x[0] in ['F', 'f']:
            return 'Female'
        elif x[0] in ['M', 'm']:
            return 'Male'
    return x

data_lab_Data2['gender'] = data_lab_Data2['gender'].apply(categorize_gender)
```

```python
def replace_state_names(df):
    replacements = {
        'WA': 'Washington',
        'Cali': 'California',
        'AZ': 'Arizona'
    }
    df['state'] = df['state'].replace(replacements)
    return df

data_lab_Data2 = replace_state_names(data_lab_Data2)
```

```python
def replace_bachelor(df):
    df['education'] = df['education'].replace('Bachelor', 'Bachelors')
    return df
data_lab_Data2 = replace_bachelor(data_lab_Data2)
```

```python
def remove_percentage_sign(df):
    df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '')
    return df

data_lab_Data2 = remove_percentage_sign(data_lab_Data2)
```

```python
def replace_vehicle_class(df):
    replacements = {
        'Sports Car': 'Luxury',
        'Luxury SUV': 'Luxury',
        'Luxury Car': 'Luxury'
    }
    df['vehicle_class'] = df['vehicle_class'].replace(replacements)
    return df

data_lab_Data2 = replace_vehicle_class(data_lab_Data2)
```

```python
for column in data_lab_Data2.columns:
    unique_values = data_lab_Data2[column].unique()
    print(f'Valeurs uniques de la colonne {column}: {unique_values}')
```
